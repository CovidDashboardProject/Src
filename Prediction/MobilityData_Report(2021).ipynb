```python
import pandas as pd
#importing dataset
df = pd.read_csv('/content/drive/MyDrive/Covid Dashboard Project/Raw Data/Maharashtra_MobilityData_2021.csv')
df
```

```python
df.drop(['country_region_code', 'country_region','metro_area',	'iso_3166_2_code',	'census_fips_code',	'place_id'], axis = 1,inplace = True)
df
```

```python
df.isnull().sum() #here 256 because it counts values of entire maharashtra
```

```python
import matplotlib.pyplot as plt 
df[0:256].plot(x='date',	y=['retail_and_recreation_percent_change_from_baseline','grocery_and_pharmacy_percent_change_from_baseline','parks_percent_change_from_baseline','transit_stations_percent_change_from_baseline',	'workplaces_percent_change_from_baseline',	'residential_percent_change_from_baseline']	,figsize = (15,8))    
plt.title("Visiting places percent change from baseline for MAHARASHTRA")
plt.xlabel("Date")
plt.ylabel("Percent change (%)") #0 denotes baseline i. e. no change
plt.show()
```

```python
import matplotlib.pyplot as plt 
df[256:].plot(x='sub_region_2',	y=['retail_and_recreation_percent_change_from_baseline','grocery_and_pharmacy_percent_change_from_baseline','parks_percent_change_from_baseline','transit_stations_percent_change_from_baseline',	'workplaces_percent_change_from_baseline',	'residential_percent_change_from_baseline']	,figsize = (15,8))    
plt.title("Visiting places percent change from baseline for Subregions")
plt.xlabel("Subregions")
plt.ylabel("Percent change (%)") #0 denotes baseline i. e. no change
plt.show()
```
